```python
# pega correlação entre g's
dados.loc[:, 'g0':'g771'].corr()
```

```python
# Plota correlações entre g's (de g0 a g50) em heatmap
# Correlação não implica necessariamente causalidade
# Correlação é um indicativo apenas para investigar melhor

import numpy as np

corr = dados.loc[:, 'g0':'g50'].corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidth=.5, cbar_kws={'shrink': .5})
```

```python
# Fazer mesmo gráfico de correlações, mas para tipos celulares (c's)
# O valor de c's na tabela significam quantas células sobreviveram ao tratamento
# Alta correlação entre tipos celulares faz sentido
# Normalmente células morrem por mesma razões, mesmo sendo de diferentes tipos

corr_celular = dados.loc[:, 'c0':'c50'].corr()

mask = np.triu(np.ones_like(corr_celular, dtype=bool))

f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidth=.5, cbar_kws={'shrink': .5})
```

Merge de dados e Análise de resultados

```python
import pandas as pd

dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados
```

```python
dados_resultados['acat_inhibitor'].unique()
```

```python
# 0 e 1 indicam se os genes foram ativados (1 -> ativado)
dados_resultados['acetylcholine_receptor_agonist'].unique()
```

```python
dados_resultados.info()
```

```python
# Soma quantas vezes cada elemento foi ativado (mostrado em ordem decrescente)
contagem_moa = dados_resultados.drop('id', axis=1).sum().sort_values(ascending=False)
contagem_moa
```

```python
# Soma valores das colunas ao invés das linhas e coloca como nova coluna
dados_resultados['a_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
dados_resultados['ativo_moa'] = (dados_resultados['a_moa'] != 0)
display(dados_resultados)
```

```python
# Junta tabela de dados com a_moa e ativo_moa
dados_combinados = pd.merge(dados, dados_resultados[['id','a_moa', 'ativo_moa']], on='id')
display(dados_combinados)
```

```python
# Verifica se existe algum caso em que o tratamento foi "com_controle" foi ativo
# Obs.: Nenhum é para ser (sem drogas não é para ter algum comportamento)
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
# Verifica se existe algum caso em que o tratamento foi "com_droga" foi ativo
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotagem de boxplot de g0 em função da droga
# Mostra quais estão ativos também
composto_principal = dados_combinados['composto'].value_counts().index[:5]
plt.figure(figsize=(12, 8))
sns.set()
sns.boxplot(data = dados_combinados.query('composto in @composto_principal'), y='g0', x='composto', hue='ativo_moa')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Machine Learning -> treinando máquina a identificar em que casos ocorre ativação de mecanismos

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state=376, stratify=y)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
dados_combinados['ativo_moa'].value_counts(normalize=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state=376, stratify=y)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(15, 10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state=376, stratify=y)


teste=[]
treino=[]

for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth = i)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append(modelo_arvore.score(x_teste, y_teste))
  treino.append(modelo_arvore.score(x_treino, y_treino))
```
